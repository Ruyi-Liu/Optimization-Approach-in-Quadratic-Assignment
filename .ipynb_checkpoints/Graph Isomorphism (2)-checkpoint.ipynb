```python
nx.draw(g,with_labels =True)
plt.savefig('A not friendly.eps', dpi=500)
```

```python
#Construct a n-degree permutation matrix 

def permu(N):
    E=np.identity(N)
    permutation = np.random.permutation(7)
    P=np.zeros([N,N]) #initialize the permutation matrix
     
    for i in range(0,N):
        P[i]=E[permutation[i]]
    
    
    return P.astype(int)

    
```

```python

np.random.seed(3)
P_1 = permu(7)
print("P=\n",P_1)


adj_g = nx.to_numpy_matrix(g)
A = adj_g
B = np.matmul(P_1.T, np.matmul(A,P_1))
print("A=\n",A)
#print(B.transpose())
print("B=\n",B)

G_2 = nx.from_numpy_matrix(B)
nx.draw(G_2,with_labels=True)
plt.savefig('B not friendly.eps', dpi=500)

print(LA.norm(P_1.T@A-B@P_1.T, 'fro') )
```

```python
# Example of cvxpy
#np.random.seed(1)
from numpy import linalg as LA

# Define and solve the CVXPY problem.
P = cp.Variable((7,7))
# print(type(P))
# arr = np.array([[1],[1],[1],[1],[1],[1],[1]])
arr = cp.expressions.constants.Constant([[1 for i in range(7)]])
constraints = [P.__matmul__(arr) == arr, P.T.__matmul__(arr) == arr]
# constraints = [np.matmul(P,arr) == arr, np.matmul(P.T,arr) == arr]
# constraints = [np.multiply(P,arr) == arr, np.multiply(P.T,arr) == arr]
prob = cp.Problem(cp.Minimize((cp.norm(P.__matmul__(A)-(P.T.__matmul__(B.T)).T, 'fro'))**2), constraints)
# prob = cp.Problem(cp.Minimize(LA.norm(np.multiply(P,A)-np.matmul(B,P), 'fro')), constraints)
prob.solve()


# Print result.
print("\nThe optimal value is", prob.value)
print("The optimal P is")
pp = pprint.PrettyPrinter(depth=6)
pp.pprint(P.value)
#print("The norm of the residual is ", cp.norm(A @ x - b, p=2).value)
```

```python

# Define and solve the CVXPY problem.
P = cp.Variable((7,7))
arr = cp.expressions.constants.Constant([[1 for i in range(7)]])
constraints = [P.__matmul__(arr) == arr, P.T.__matmul__(arr) == arr, cp.min(P)>=0]
prob = cp.Problem(cp.Minimize((cp.norm(P.__matmul__(A)-(P.T.__matmul__(B.T)).T, 'fro'))**2), constraints)
# prob = cp.Problem(cp.Minimize(LA.norm(np.multiply(P,A)-np.matmul(B,P), 'fro')), constraints)
prob.solve()


# Print result.
print("The norm of the residual is", prob.value)
print("The optimal P is")

p = []
c=0
for i in P.value:
    p.append([])
    for j in i:
        p[c].append(round(j, 2))
    c+=1
                 
print(p)       
        
print(p == P_1.T)
print(P_1.T)

```

```python
# Check Friendliness
def Friendly(A):
    w, v = LA.eig(A) 
    
    w_round = []
    for i in w:
        w_round.append(round(i,15))
    W = set(w_round)
    
    if len(W)!=len(A):
        print("Not Friendly")
        print("Eigenvalues not distinct")
        return False
        
    for i in range(len(v)):
        s = np.sum(v[i])
        if abs(s) < 10**(-20):
            print("Eigenvector sum up to 0")
            print("Not Friendly")
            return False
        
    print("\nFriendly")
    return True
```
